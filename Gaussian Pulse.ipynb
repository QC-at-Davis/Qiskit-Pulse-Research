## Pulse
* Go from low to high signal, then back to low
* pulse "shape" arises from *pulse-shaping* 

## Gaussian Pulse
* shaped as a gaussian function, produced with a gaussian filter

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(-1, 1, 2 * 100, endpoint=False)
i, q, e = signal.gausspulse(t, fc=5, retquad=True, retenv=True)
plt.plot(t, i, t, q, t, e, '--')
```
